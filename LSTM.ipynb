```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM, Dense
```

```python
# 1 Load dataset(top 10,000 words only)
(X_train,y_train),(X_test,y_test) = imdb.load_data(num_words=10000)
```

```python
# 2. Pad sequences (make all reviews same length)
max_len = 100
X_train = pad_sequences(X_train,maxlen=max_len)
X_test = pad_sequences(X_test,maxlen=max_len)
```

```python
#3.Build model
model = Sequential([
    Embedding(input_dim=10000,output_dim=32,input_length=max_len),  #output_dim = vector size ko reduce karke 32 kar diya(of input_dim )
    LSTM(units=32),
    Dense(units=1,activation='sigmoid') #because it is binary classification so we use sigmoid
])
```

```python
# 4.Compile
model.compile(
    optimizer ='adam',
    loss  = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
#5 Train
model.fit(
    X_train,
    y_train,
    batch_size=64,
    epochs=3,

)
```

```python
#6. Evaluate
loss,accuracy = model.evaluate(X_test,y_test)
print("Accuracy:",accuracy)
```

```python
#7.Decode reviews(numbers back to words)
word_index = imdb.get_word_index()
reverse_word_index = {value:key for key, value in word_index.items()}

def decode_review(encoded_review):
  return ' '.join([reverse_word_index.get(i - 3,'?') for i in encoded_review if i >= 3])

```

```python
#8. predict on test data
predictions = model.predict(X_test)
predicted_labels = (predictions > 0.5).astype(int).flatten()
```

```python
# 9. Show some correct predictions
print("\nCORRECT PREDICTIONS:\n")
count = 0
for i in range(len(X_test)):
    if predicted_labels[i] == y_test[i]:
        print("Review:", decode_review(X_test[i])[:300], "...")
        print("Actual:", y_test[i], "| Predicted:", predicted_labels[i], "| Confidence:", predictions[i][0])
        print("-" * 80)
        count += 1
    if count == 2:
        break
```

```python
# 10. Show some wrong predictions
print("\nWRONG PREDICTIONS:\n")
count = 0
for i in range(len(X_test)):
    if predicted_labels[i] != y_test[i]:
        print("Review:", decode_review(X_test[i])[:300], "...")
        print("Actual:", y_test[i], "| Predicted:", predicted_labels[i], "| Confidence:", predictions[i][0])
        print("-" * 80)
        count += 1
    if count == 3:
        break
```
