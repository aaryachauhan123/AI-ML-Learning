```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Input,Conv1D,GlobalMaxPooling1D,Dense
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
X = np.array([[[3],[3],[3],[3],[3],[3],[3]],
              [[2],[2],[2],[2],[2],[2],[2]],
              [[7],[7],[7],[7],[7],[7],[7]],
              [[8],[8],[8],[8],[8],[8],[8]]],dtype ='float32')

#Labels:1 = effcient (low emission/low usage),
y = np.array([1,1,0,0],dtype = 'int32') # why we take this 0 and 1 value?

print("X shape",X.shape)
print("y shape",y.shape)
print("y ",y)
```

```python
# split the data
X_train,X_test,y_test,y_train = train_test_split(X,y,test_size = 0.5,random_state=42)
print("X_train shape",X_train.shape)
print("X_test shape",X_test.shape)
```

```python
# build the model
model = Sequential([  #CNN model
    Input(shape = (7,1)),
    Conv1D(4,3,activation='relu'),
    GlobalMaxPooling1D(),
    Dense(1,activation='sigmoid')
])
```

```python
#compile with accuracy metric
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
    )
```

```python
#train
history = model.fit(
    X_train,
    y_train,
    verbose = 1,  #the output created because of this line
    epochs=100,
)
print("training complete")
```

```python
#step 1: predict probability on test data
y_prob=model.predict(X_test,verbose=0) #x test pe lagta hai yai
print("Predicted probabilites:",y_prob)

#step 2:
threshold = 0.5
y_pred = (y_prob >= threshold).astype(int)
print("predicted class:",y_pred)
print("true class:",y_test)

#step 3 accuracy
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)

#classification report
cr=classification_report(y_test,y_pred, target_names=['Not effcient','effcient'])
print(cr)

#confusion matrix
print(confusion_matrix(y_test,y_pred))
```
