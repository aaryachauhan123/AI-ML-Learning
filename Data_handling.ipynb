```python
import pandas as pd
import numpy as np

data = {
    "Energy source" : ["solar","wind","Hydropower","Geothermal","Biomass","Nuclear"],
    "Energy consumption" :[1200,np.nan,2900,np.nan,2500,3200],
    "cost(million$)" :[200,400,np.nan,150,250,np.nan]
}

energy_df=pd.DataFrame(data);
print(energy_df)



```

```python
#forward fill
forward_filled_df = energy_df.fillna(method="ffill")
print("\ndata after forward fill")
print(forward_filled_df)
```

```python
#backword fill
backword_filled_df = energy_df.fillna(method="bfill")
print("\ndata after backword fill")
print(backword_filled_df)
```

use of isna() function

```python
#checking for null valus using isna() function
#
energy_df["missing consumption"] = energy_df["Energy consumption"].isna();
print(energy_df)
print("\n")
energy_df["missing consumption"] = energy_df["Energy consumption"].isna().astype(int);
print(energy_df)
```

```python
#for cost clm using isna()
energy_df["missing cost"] = energy_df["cost(million$)"].isna();
print(energy_df)
print("\n")
energy_df["missing cost"] = energy_df["cost(million$)"].isna().astype(int);
print(energy_df)
```

use of min max scaler formula

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#min max scaler method using to convert the large numbers between 0 to 1
scaler = MinMaxScaler()

energy_df[["Energy consumption","cost(million$)"]]=scaler.fit_transform(
    energy_df[["Energy consumption","cost(million$)"]]
)
print(energy_df)
```

Standardization method


```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

energy_df[["Energy consumption","cost(million$)"]]=scaler.fit_transform(
    energy_df[["Energy consumption","cost(million$)"]]
)
print("\ndata after standardization (z score scalling)")
print(energy_df)
```

Encoding categorial variables


```python
energy_encoded_df = pd.get_dummies(energy_df,columns=["Energy source"])
print("\nData after one hot encoding categorial variable")
print(energy_encoded_df)
```

```python
#using astype to show 0 and 1 instead false and true

#for encoding in data there will be no null values in dataframe
#if there is null valus we can use mean() to remove value with mean value.(not backword or forward filling)

energy_encoded_df = pd.get_dummies(energy_df,columns=["Energy source"]).astype(int)
print("\nData after one hot encoding categorial variable")
print(energy_encoded_df)
```

Data cleaning

```python
# removing null values
cleaned_df= energy_df.dropna()
print("\n")
print(cleaned_df)
```

```python
#replacing null value in energy consumption clm with mean

#energy_df["Energy consumption"].fillna(energy_df["Energy consumption"].mean()) # will not change in original database

energy_df["Energy consumption"].fillna(energy_df["Energy consumption"].mean(),inplace=True) # will  change in original database

energy_df["cost(million$)"].fillna(energy_df["cost(million$)"].mean(),inplace=True) #to changes in original dataset

print("\ndata after imputing null values with mean")
print(energy_df)
```
