```python
import matplotlib.pyplot as plt
```

```python
# sample data for energy consumption over 6 months
months = ['jan','feb','Mar','apr','May','Jun']
energy_consumption = [1200,1300,1100,1500,1400,1600]
#create line plot
plt.plot(months,energy_consumption,marker='o',color='r',linestyle='--')
plt.title('Energy consumption over 6 months')
plt.xlabel('Months')
plt.ylabel('Energy consumption(Mwh)')
plt.show()

# marker='o' means jo data points he who point me dikhenge
```

Bar graph

```python
plt.bar(months,energy_consumption,color='y')

#plt.bar(energy_consumption,months,color='y')
```

PIE chart

```python
plt.pie(energy_consumption,labels=months,autopct='%1.1f%%',colors=['red','pink','blue','gold','green','yellow'],explode=(0,0,0.2,0,0,0.2))

#pehle numerical argument hogi fir label me column name(non numeric), autopct is for showing values in chart using %1.1f% only the extra % is used to show in the chart
#colors is a list
```

Scatter graph

```python
plt.scatter(months,energy_consumption,color='y')
#argument just like bar graph
```

Coustmized bar chart

```python
plt.bar(months,energy_consumption,color='y',edgecolor='black',linewidth=3)

plt.title('Energy consumption over 6 months')
plt.xlabel('Months',fontsize=14,fontweight='bold')
plt.ylabel('Energy consumption(Mwh)',fontsize=12,fontweight='bold')
plt.grid(True,linestyle='--',alpha=0.6,color='g') #alpha means thickness of grid line
plt.legend(['energy_consumption'], loc='best') #loc means location
#plt.grid(False,alpha=0)   #if we don't want grid lines,but want in for future
plt.show()
```

```python
import seaborn as sns
import pandas as pd

#Create a Dataframe for pair plot

data = {
    "solar" : [1200,1500,3000,1800,2300],
    "wind" : [1000,2000,3000,1100,4000],
    "hydropower" :[2300,1700,1800,1200,1399],
    "Biomass":[1200,1300,1100,1500,1400,]
}

energy_df = pd.DataFrame(data)
sns.pairplot(energy_df)
plt.show()
```

heat map

```python
correlation_matrix = energy_df.corr() # to make correlation matrix
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',linewidth=0.5)
plt.title('correlation between energy source',fontsize=10,fontweight='bold')
```

heat map in diffrent color

```python
sns.heatmap(correlation_matrix,annot=True,cmap='Purples',linewidth=0.5)
```
